# Afri Align
Responsible AI Assessment & Guidance Tool for African Contexts

Afri Align is an LLM-powered web tool that helps organizations, governments, and developers in Africa assess whether their AI projects are ethical, sustainable, and human-centered. The user inputs their AI project details and the tool score it across five responsible AI pillars, flags risks specific to low-resource and developing-country contexts, and generates a tailored policy and action roadmap as a downloadable report.

## The Five Assessment Pillars
1. Ethics & Fairness	
Bias in training data, representation, consent, algorithmic harm
2. Governance & Accountability	
Who owns the AI? Who is liable? What oversight exists?
3. Human Impact	
Does it serve or displace communities? Whose needs does it centre?
4. Sustainability	
Energy use, long-term maintainability, local infrastructure fit
5. Context Sensitivity	
Local language support, infrastructure limits, policy gaps, digital divide


## User flow of the App

Step 1 Landing page
User reads a one-paragraph description of AFRI Align and clicks 'Start Assessment'

Step 2 Project intake form (5 questions, one per pillar)
Q1 Ethics:       What data does your AI use? Who collected it? Was consent obtained?
Q2 Governance:   Who owns this system? What happens if it causes harm?
Q3 Human Impact: Which communities does this affect? Were they consulted?
Q4 Sustainability: What infrastructure does it require? Who maintains it long-term?
Q5 Context:      Does it work in low-bandwidth settings? Which languages does it support?

Step 3 — Scoring function  (your JavaScript rubric)
Each answer is scored 0–4 per pillar
Total score = sum across 5 pillars (max 20)
Risk tier = Low (16–20) / Moderate (10–15) / High (0–9)

Step 4 — Claude API call
Send: project description + scores + risk tier
Receive: narrative analysis + 5 specific recommendations + policy roadmap

Step 5 — Report page
Display: pillar scores, risk tier badge, Claude's narrative, recommendations
Button: 'Download Report' (Markdown or print-to-PDF)
